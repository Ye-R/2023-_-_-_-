```python
import numpy as np
import pandas as pd
import natsort
import os, glob

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "arial"
```

### raw data plotting, cropping

```python
file_list = glob.glob(f'./odor_data/*/*.csv')
natsort.natsorted(file_list)
```

```python
EXP = 12 # 
# odorlist = ['Mint', 'Orange', 'Smoked', 'BBQ','Ethanol', 'LighterOil', 'Clorox'] #  ()
odorlist = ['Ethanol', 'Smoked','Mint'] #  ()
sensorlist = ['NH3', 'C2H4', 'H2', 'CH2O', 'VOC','Humidity','Temperature']
dt_length = 100 #   


for exp_no in range(1, EXP+1):
    if not os.path.isdir('./odor_data_cropped'):
        os.mkdir(f'./odor_data_cropped')

    if not os.path.isdir(f'./odor_data_cropped/odor_data{exp_no}'):
        os.mkdir(f'./odor_data_cropped/odor_data{exp_no}')
    
    for odor in odorlist:
        dirlist = natsort.natsorted(glob.glob(f'./odor_data/odor_data{exp_no}/{odor}_*.csv'))
        if(len(dirlist) == 0):
            continue
        
        # ,  
        dt_target = pd.DataFrame([])
        lasttime = 0

        for dir_ in dirlist:
            dt = pd.read_csv(dir_)
            #  
            dt['Time'] += lasttime
            lasttime = np.array(dt['Time'])[-1]
            #   
            dt_target = pd.concat([dt_target, dt], axis=0)
            #     
            dt_target[['NH3', 'C2H4', 'H2']] = dt_target[['NH3', 'C2H4', 'H2']] + (dt_target[['NH3', 'C2H4', 'H2']] <= 50).astype(int) * 255
            #  
            dt_target = dt_target[dt_target['CO2'] <= 5000]
            dt_target = dt_target[dt_target['VOC'] <= 3]
            dt_target = dt_target[dt_target['CH2O'] <= 2000]
            dt_target = dt_target[dt_target['Humidity'] <= 100]
            dt_target = dt_target[dt_target['Temperature'] <= 100]
            dt_target = dt_target[dt_target['PM2.5'] <= 1000]
            dt_target = dt_target[dt_target['PM10'] <= 1000]
            dt_target = dt_target[dt_target['PM1.0'] <= 1000]
            
        print(f'Experiment {exp_no} odor {odor} shape: {dt_target.shape}') # 
        
        #   
        fig, ax = plt.subplots(len(sensorlist),1, figsize=(10, len(sensorlist)*1.5))
        for idx, ss in enumerate(sensorlist):
            ax[idx].plot(dt_target['Time'], dt_target[ss], c=[0+0.15*idx,1-0.15*idx,0.3+0.1*idx,0.5], lw=1, label = ss);
            ax[idx].legend();
        ax[idx].set_xlabel('Time(sec)');
        ax[0].set_ylim(100,280); ax[1].set_ylim(100,280); ax[2].set_ylim(100,280); ax[4].set_ylim(-1,4);
        plt.tight_layout()
        plt.show()

        for i in range(len(dt_target)//dt_length):
            odor_data_cropped = dt_target.iloc[dt_length*i:dt_length*(i+1),:]
            odor_data_cropped.to_csv(f'./odor_data_cropped/odor_data{exp_no}/{odor}_{i}.csv', index=None)
            
        print(f'Experiment {exp_no} odor {odor} no. of cropped data: {i}') # 
        print('\n\n')
```
